```python
# Minimal Jupyter Notebook example:
  
# Save plot to file only, then display the saved image
import os
import matplotlib.pyplot as plt
from IPython.display import Image, display

# Create plots directory if needed
os.makedirs("plots", exist_ok=True)

# Example data
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 5, 3]
# Add a comment

# Create figure
plt.figure()
plt.plot(x, y, marker='o')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Simple Data Plot")

# Save figure to file
filename = "plots/simple_plot.png"
plt.savefig(filename, dpi=300, bbox_inches="tight")

# Close figure so it is NOT stored as notebook output
plt.close()

# Display saved file in notebook
display(Image(filename=filename))
```
